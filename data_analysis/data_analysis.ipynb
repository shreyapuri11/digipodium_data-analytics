# Data Manipulation

 - Adding columns

```python
# adding a total column
years= list(range(1980,2014))
canada['Total']=canada[years].sum(axis=1) # axis=1 means sum across columns(horizontally)
canada.head()
```

```python
canada['Faltu'] = 10
canada.head()
```

Removing a column

```python
# single drop column

canada.drop(columns='Faltu',inplace=True)
```

```python
canada.head()
```

```python
# multiple columns drop
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
canada.drop(columns=cols_to_drop,inplace=True)
canada.head()
```

Renaming columns

```python
canada.rename(mapper={
    'AreaName': 'Continent',
    'RegName':'Region',
    'DevName': 'Status'
}, axis=1,inplace=True)
canada.head(4)
```

Replacing data

```python
canada.Status.unique()
```

```python
canada.replace('Developing regions',0,inplace=True)
canada.replace('Developed regions',1,inplace=True)
canada.head()
```

Applying lambda operation on column

```python
vs= canada['Total'].apply(lambda x: x>50000).value_counts()
vs.plot(kind='bar')
```

```python
def categorize_by_weight(value):
    if value == np.nan:
        return np.nan
    elif value < 3900:
        return  'Light Weight'
    elif value >= 3900 and value < 5100:
        return 'Medium Weight'
    elif value >= 5100:
        return "Heavy Weight"
categorize_by_weight(3000)
```
