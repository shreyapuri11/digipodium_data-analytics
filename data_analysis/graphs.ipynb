# Visualiztion on iris dataset

```python
iris = sns.load_dataset('iris') # about iris flower
iris
```

- single column with numbers
  - stats
    - histogram
    - boxplot
    - density plot (KDE) with the help of pandas
  - discrete
    - line plot
    - area chart
    - bar graph
- 2 or 3 columns with numbers
  - relationship
    - scatter plot
    - bubble chart(scatter plot with sizes)
    - heat map
    - hexagonal bin plot
  - composition
    - stacked bar chart
    - stacked area chart
    - grouped line chart
  

```python
_ = plt.hist('sepal_length',data = iris, bins =20) # ranges are decided by bins in histogram
```

```python
_ = plt.hist('petal_length',data = iris, bins=20)
```

```python
fig, axs = plt.subplots(nrows = 2, ncols =2, figsize=(5,5), layout = 'constrained')
axs[0,0].hist('sepal_width',data=iris,bins = 20)
axs[0,1].hist('petal_width',data=iris,bins = 20)
axs[1,0].hist('sepal_length',data=iris,bins = 20)
axs[1,1].hist('petal_length',data=iris,bins = 20)
axs[0,0].set_title('sepal_width')
axs[0,1].set_title('petal_width')
axs[1,0].set_title('sepal_length')
axs[1,1].set_title('petal_length')
plt.show()

```

# Visualiztion on penguin dataset

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

```python
_ = plt.hist('bill_length_mm',data = penguins, bins = 15,color='green')
plt.title('Histogram of bill_length_mm')
plt.show()
```

```python
plt.hist('bill_depth_mm',data = penguins,bins = 10,color = 'blue')
plt.title("Histogram of bill_depth_mm")
plt.grid = True
plt.show()
```

```python
_ = plt.hist('flipper_length_mm',data = penguins,bins = 20,color='orange')
plt.title('Histogram of flipper_length_mm')
plt.grid = True
plt.show()
```

```python
plt.hist('body_mass_g',data = penguins,bins = 20,color='magenta')
plt.title("Histogram of body_mass_g")
plt.grid= True
plt.show()
```

```python
fig, axs = plt.subplots(nrows = 2, ncols =2, figsize=(5,5), layout = 'constrained')
axs[0,0].hist('bill_length_mm',data=penguins,bins = 20,color='red')
axs[0,1].hist('bill_depth_mm',data=penguins,bins = 20,color='green')
axs[1,0].hist('flipper_length_mm',data=penguins,bins = 20,color='orange')
axs[1,1].hist('body_mass_g',data=penguins,bins = 20,color='blue')
axs[0,0].set_title('bill_length_mm')
axs[0,1].set_title('bill_depth_mm')
axs[1,0].set_title('flipper_length_mm')
axs[1,1].set_title('body_mass_g')
plt.legend()
plt.show()
```

```python
plt.boxplot('body_mass_g',data =penguins,notch=True)
plt.show()
```

```python
plt.plot('bill_length_mm',data = penguins,color='blue')
plt.plot('bill_depth_mm',data = penguins,color='red')
plt.legend()
plt.show()
```

```python
plt.plot('body_mass_g',data = penguins,color='blue')
plt.legend()
plt.show()
```

```python
plt.plot('flipper_length_mm',data = penguins,color='blue',marker='o',mfc='red',mec='black')
plt.legend()
plt.show()
```

```python
plt.bar(x=penguins.index,height='bill_length_mm', data=penguins,color='orange',width= 0.5,bottom = 10)
plt.show()
```

```python
plt.scatter(x='bill_length_mm',y='bill_depth_mm',data = penguins,c='b',marker='*')
plt.show()
```

```python
plt.scatter(x='bill_depth_mm',y='bill_length_mm',data = penguins,c=penguins.index)
plt.colorbar()
plt.show()
```

```python
def convert_to_number(name):
    if name == 'Male':
        return 1
    if name == 'Female':
        return 2
    else:
        return 3
penguins['s']= penguins['sex'].apply(convert_to_number)
penguins
```

```python
plt.scatter(x='flipper_length_mm', y='body_mass_g',data=penguins, c='s',cmap ='rainbow')
plt.colorbar()
plt.title('flipper_length_mm Vs body_mass_g')
plt.show()
```

```python
plt.scatter(x='bill_depth_mm', y='bill_length_mm',data=penguins, c='s',cmap ='cool')
plt.title('bill_depth_mm Vs bill_length_mm')
plt.show()
```

```python
def convert_to_number(name):
    if name == 'Adelie':
        return 1
    if name == 'Gentoo':
        return 2
    else:
        return 3
penguins['spc']= penguins['species'].apply(convert_to_number)
penguins
```

```python
plt.scatter(x='bill_depth_mm', y='bill_length_mm',data=penguins, c='spc',cmap ='viridis',marker='^')
plt.colorbar()
plt.grid= True
plt.show()
```
