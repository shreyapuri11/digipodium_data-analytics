```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
print(titanic['survived'].unique()) #categorical columns
print(titanic['pclass'].unique())
print(titanic['sex'].unique())
```

# Categorical graphs
 - columns that have text are generally categorical
 - the first step is to count the number of unique values in column
 - then plot the count of each category
 - folowing charts are a good way to visualize categorcal data
    - bar chart - 
    - pie chart
    - donut chart
    - line chart - when too many categories
    - area chart - when too many categories

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind = 'pie')
```

```python
survival_count = titanic['survived'].value_counts()
survival_count.plot(kind = 'pie', autopct='%.1f%%')
```

```python
survival_count = titanic['alive'].value_counts()
survival_count.plot(kind = 'pie', autopct='%.1f%%',title='Survival rate')
```

```python
p_class = titanic['pclass'].value_counts()
p_class.plot(kind = 'pie', autopct='%.1f%%')
```

```python
survival_count.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'Survival Rate',
    labels = ['Survived','Not Survived'],
    colors = ['#66b3ff','#ff9999'],
    labeldistance=1.1,
    wedgeprops={
        'width' : 0.5,
    })
```

```python
p_class = titanic['pclass'].value_counts()
p_class
```

```python
p_class.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'Number of peolpe in different levels of ships',
    labels = ['3','1','2'],
    colors = ['#66b3ff','#ff9999','#ffbb11'],
    labeldistance=1.1,
    wedgeprops={
        'width' : 0.5,
    })
```
