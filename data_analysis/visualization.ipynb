```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 
```

```python
fig, ax = plt.subplots(figsize=(5,2))
x = [1,2,3,4,5]
y = [3,4,4,5,3]
ax.plot(x,y,'o-')
```

```python
fig = plt.figure(figsize=(3,3))
plt.plot(x,y)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
plt.show()
```

```python
x= np.arange(-5,5,0.2)
y1 = np.sin(x)
y2 = np.cos(x)
fig,axs = plt.subplots(nrows = 1, ncols = 2, figsize =(10,3))
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
plt.show()
```

```python
x= np.arange(-5,5,0.2)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
fig,axs = plt.subplots(nrows = 1, ncols = 3, figsize =(10,3))
axs[0].plot(x,y1,'r--')
axs[1].plot(x,y2,'b--')
axs[2].plot(x,y3,'g--')
axs[0].set_title('sin(x)')
axs[1].set_title('cos(x)')
axs[2].set_title('tan(x)')
plt.show()
```

# Implicit vs Explicit graphs

```python
# plt -> implicit figure and axes creation
x = np.linspace(-10,10,25)
plt.plot(x,x,'-',label = 'linear')
plt.plot(x,x**2,'--',label = 'quadratic')
plt.plot(x,x**3,'o:',label='cubic')
plt.xlabel('my x label')
plt.ylabel('my y label')
plt.title('Simple Plot with implicit graph setting')
plt.legend()
plt.show()
```
